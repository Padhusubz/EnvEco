```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters for MRS curve
income_range = np.linspace(1000, 50000, 400)  # Income range from $1,000 to $50,000
initial_mrs = 500  # Initial MRS value
decay_rate = 0.0001  # Decay rate for exponential function

# Calculate the MRS curve using an exponential decay function
mrs_curve = initial_mrs * np.exp(-decay_rate * income_range)

# Define the function for indifference curves
def indifference_curve(income, pollution, happiness_level, a, b, c):
    return (happiness_level - a - b * income) / c

# Define parameters for the linear happiness function H = a + bI + cP
a = 0  # Intercept
b = 0.045 / 10000  # Coefficient for income
c = -0.015 / 10  # Coefficient for pollution

# Define a range of pollution values
pollution_range = np.linspace(0, 100, 400)  # Pollution range

# Define happiness levels for the indifference curves
happiness_levels = [1, 2, 3]

# Plot the MRS curve
plt.figure(figsize=(10, 6))
plt.plot(income_range, mrs_curve, label='Diminishing MRS Curve', color='blue')

# Plot the indifference curves
for happiness_level in happiness_levels:
    indiff_curve = indifference_curve(income_range, pollution_range, happiness_level, a, b, c)
    plt.plot(income_range, indiff_curve, label=f'Indifference Curve for H={happiness_level}')

# Labels and title
plt.xlabel('Income ($)')
plt.ylabel('Pollution Level (μg/m³)')
plt.title('MRS and Indifference Curves')
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()

```
