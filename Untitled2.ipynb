```python
import matplotlib.pyplot as plt

# Data points
income_loss_due_to_drought = 1000  # Actual income loss due to decline in agricultural yields (A$ per person-year)
total_marginal_value = 18000  # TMV of drought conditions (A$ per household-year)
marginal_value_income = 17000  # MV of income loss (A$ per household-year)

# X-axis labels
categories = ['Income Loss Due to Drought', 'Total Marginal Value']

# Y-axis values
values = [income_loss_due_to_drought, total_marginal_value]

# Bar plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=['blue', 'orange'])

# Adding MV of income loss as a separate point
plt.scatter('Income Loss Due to Drought', marginal_value_income, color='red', zorder=5)
plt.text('Income Loss Due to Drought', marginal_value_income + 500, 'MV of Income Loss', ha='center', color='red')

# Adding values on top of the bars
for i, v in enumerate(values):
    plt.text(i, v + 500, f'A${v}', ha='center')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Value (A$ per household-year)')
plt.title('Impact of Drought Conditions on Income and Happiness')
plt.grid(True)
plt.show()
```
